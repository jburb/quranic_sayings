```python
with open('files/en.arberry_say.json') as f:
    arberry_sayings = json.load(f)
```

```python
arberry_sayings_flattened = pd.json_normalize(arberry_sayings['data']['matches'])
```

```python
arberry_sayings_flattened.columns = arberry_sayings_flattened.columns.str.replace('.', '_') # replace flattend colsnames' dots with underscores
```

```python
theoretical_muhammad_pbuh_sayings = arberry_sayings_flattened.query('text.str.contains("Say: \'")')
```
